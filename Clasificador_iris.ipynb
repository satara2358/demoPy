```python
from sklearn import datasets
iris = datasets.load_iris()

dir(iris)
```

```python
print(iris.DESCR)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
iris.data
```

```python
iris.target
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['specie'] = iris.target
df.head()
```

```python
sns.pairplot(df, hue='specie')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
knn = KNeighborsClassifier(n_neighbors=3)

X = iris.data
Y = iris.target

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

knn.fit(X_train, Y_train)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
Y_pred = knn.predict(X_test)
mc = confusion_matrix(Y_test, Y_pred)
sns.heatmap(mc, annot=True)
plt.show()
```

```python
print(f'es tan exacto que {accuracy_score(Y_test, Y_pred)}' )
```

```python
import gradio as gr
def saludar(nombre):
  return "Hola " + nombre

demo = gr.Interface(fn=saludar, inputs="text", outputs="text")
demo.launch()

# Gradio.app docs
```

```python
## 01 setosa 02 versicolos 03 virginica
def model(sepal_length, sepal_width, petal_length, petal_width):
  specie = ['Iris-Setosa','Iris-Versicolor', 'Iris-Virginica']

  index = knn.predict([[sepal_length, sepal_width, petal_length, petal_width]])[0]
  return specie[index]

interfaz = gr.Interface(
    fn=model,
    inputs=[
        gr.Number(label='Sepal Length'),
        gr.Number(label='Sepal Width'),
        gr.Number(label='Petal Length'),
        gr.Number(label='Petal Width')
    ],
    outputs=gr.Textbox(label='Species'),
    examples=[[7.9, 6.1, 5.5, 5,9], [6.1, 0.2, 0.4, 0.9], [3.3, 4.1, 2.2, 3.1]],
    title='Iris Flower Classifier',
    description="modelos de prueba clasificador de flores",
    article="app demo",
    theme='dark'
)

interfaz.launch()

#df.head()
```

```python
print(model(5.1, 3.2, 5.1, 5.4))
```
